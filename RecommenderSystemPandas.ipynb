### Importing necessary Libraries

```python
import pandas as pd
import numpy as np
from math import sqrt
import matplotlib.pyplot as plt
%matplotlib inline
```

### Gettting the dataset from MovieLens

```python
movies_df = pd.read_csv("movieslist.csv")
movies_df.head()
```

### Extracting the year from title and creating a new column

```python
movies_df['year'] = movies_df.title.str.extract('(\(\d\d\d\d\))',expand = False)
movies_df['year'] = movies_df.year.str.extract('(\d\d\d\d)',expand = False)
movies_df['title'] = movies_df.title.str.replace('(\(\d\d\d\d\))','')
movies_df['title'] = movies_df['title'].apply(lambda x: x.strip())
movies_df.head()
```

### Taking the genres and creating a List for each Movie

```python
movies_df['genres'] = movies_df.genres.str.split('|')
movies_df.head()
```

### Creating a column for each genre and filling it with 1/0

```python
moviesgenres_df = movies_df.copy()
for i, r in movies_df.iterrows():
    for j in r['genres']:
        moviesgenres_df.at[i,j] = 1
moviesgenres_df = moviesgenres_df.fillna(0)
moviesgenres_df.head()
```

### Getting the user input and creating User Profile Vector

```python
userinput = [
    {'title' : 'Spider-Man', 'rating':5},
    {'title' : 'Batman: Gotham Knight', 'rating':4.5},
    {'title' : 'Jumanji', 'rating':2},
    {'title' : 'Incredible Hulk, The', 'rating':3.5},
    {'title' : 'Iron Man', 'rating':4}
]
inputmovies = pd.DataFrame(userinput)
inputmovies
```

### Adding the column MovieId for easy data retrieval

```python
inputid = movies_df[movies_df['title'].isin(inputmovies['title'].tolist())]
inputmovies = pd.merge(inputid, inputmovies)
inputmovies = inputmovies.drop('genres',1).drop('year',1)
inputmovies
```

```python
usermovies = moviesgenres_df[moviesgenres_df['movieId'].isin(inputmovies['movieId'].tolist())]
usermovies
```

```python
usermovies = usermovies.reset_index(drop = True)
```

### Getting the Genre Table for user input movies

```python
usertable = usermovies.drop('movieId',1).drop('title',1).drop('genres',1).drop('year',1)
usertable
```

```python
inputmovies['rating']
```

### Calculating the overall genre rating of user - {USER PROFILE VECTOR}

```python
userprofile = usertable.transpose().dot(inputmovies['rating'])
userprofile
```

```python
genretable = moviesgenres_df.set_index(moviesgenres_df['movieId'])
genretable = genretable.drop('movieId',1).drop('title',1).drop('genres',1).drop('year',1)
genretable.head(100)
```

### Creating the recommendion list by finding the weighted average of genre table and user profile vector

```python
recommendtable_df = ((genretable * userprofile).sum(axis = 1))/(userprofile.sum())
recommendtable_df.head()
```

```python
recommendtable_df = recommendtable_df.sort_values(ascending = False)
recommendtable_df.head()
```
